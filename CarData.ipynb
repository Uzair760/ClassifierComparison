```python
import pandas as pd
```

```python
df = pd.read_csv('Automobile.csv')
df.isnull().any()
```

```python
df.head()
```

```python
df["name"] = df["name"].str.title()
df.head()
```

```python
in3_to_l = 0.0163871
df['displacement_litres'] = df['displacement'] * in3_to_l
df['displacement_litres'] = df['displacement_litres'].round(3)
df.head()
```

```python
lbs_to_kg = 0.453592
df['weight_kg'] = df['weight'] * lbs_to_kg
df['weight_kg'] = df['weight_kg'].round()
df.head()
```

```python
df['horsepower'] = df['horsepower'].fillna(0)
df['hp_per_litre'] = df['horsepower'] / df['displacement_litres']
df.head()
```

```python
df['hp_per_ton'] = df['horsepower'] / df['weight_kg']
df.head()
```

```python
df = df.drop(['displacement', 'weight'], axis=1)
df.head()
```

```python
df['origin'] = df['origin'].str.title()
df['origin'] = df['origin'].str.replace('Usa', 'USA')
df.head()
```

```python
df[['make', 'model']] = df['name'].str.split(' ', 1, expand=True)
df.head()
```

```python
df = df.drop('name', axis=1)
df.head()
```

```python
df.columns.values
```

```python
col_order = ['make', 'model', 'origin', 'mpg', 'cylinders', 'displacement_litres', 'weight_kg', 
             'horsepower', 'acceleration', 'model_year', 'hp_per_litre', 'hp_per_ton']
df = df[col_order]
df.head()
```

```python
df = df.rename(columns={'acceleration': '0-60 mph'})
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X = df.drop(['mpg', 'make', 'model', 'origin'], axis=1)
Y = df.mpg
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
lr_model = linear_model.LinearRegression()
lr_model.fit(X_train, Y_train)
```

```python
Y_pred = lr_model.predict(X_test)
print(Y_pred)
```

```python
coef = lr_model.coef_
intercept = lr_model.intercept_
mse = mean_squared_error(Y_test, Y_pred)
r2_score = r2_score(Y_test, Y_pred)
print('Coefficients:', coef)
print('Intercept:', intercept)
print('Mean squared error (MSE): %.2f'
      % mse)
print('Coefficient of determination (R^2): %.2f'
      % r2_score)
```

```python
df['origin'] = df['origin'].map({'USA': 1})
df['origin'] = df['origin'].fillna(0)                                 
df
```

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.linear_model import SGDClassifier
```

```python
X = df.drop(['origin', 'make', 'model'], axis=1)
Y = df.origin
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
names = ["Nearest_Neighbors", "Linear_SVM", "Polynomial_SVM", "RBF_SVM", "Gaussian_Process",
         "Gradient_Boosting", "Decision_Tree", "Extra_Trees", "Random_Forest", "Neural_Net", "AdaBoost",
         "Naive_Bayes", "QDA", "SGD"]

classifiers = [
    KNeighborsClassifier(3),
    SVC(kernel="linear", C=0.025),
    SVC(kernel="poly", degree=3, C=0.025),
    SVC(kernel="rbf", C=1, gamma=2),
    GaussianProcessClassifier(1.0 * RBF(1.0)),
    GradientBoostingClassifier(n_estimators=100, learning_rate=1.0),
    DecisionTreeClassifier(max_depth=5),
    ExtraTreesClassifier(n_estimators=10, min_samples_split=2),
    RandomForestClassifier(max_depth=5, n_estimators=100),
    MLPClassifier(alpha=1, max_iter=1000),
    AdaBoostClassifier(n_estimators=100),
    GaussianNB(),
    QuadraticDiscriminantAnalysis(),
    SGDClassifier(loss="hinge", penalty="l2")]

clf_scores = []
```

```python
for name, clf in zip(names, classifiers):
    clf.fit(X_train, Y_train)
    score = clf.score(X_test, Y_test)
    clf_scores.append(score)
```

```python
clf_scores
```

```python
df = pd.DataFrame()
df['classifier'] = names
df['score'] = clf_scores
df = df.sort_values(by='score', ascending=False)
df
```

```python
import seaborn as sns

df = pd.DataFrame()
df['classifier'] = names
df['score'] = clf_scores
sns.set(style="whitegrid")
ax = sns.barplot(y="classifier", x="score", data=df, order=df.sort_values(by='score', ascending=False).classifier)
```
